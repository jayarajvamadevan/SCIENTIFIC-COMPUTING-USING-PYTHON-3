```python
# Experiment 5.1 Solution of Ordinary Differential Equations

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
model = lambda x,t:-2*x
# initial condition
x0 = 1
# time points
t = np.linspace(0,20)
# solve ODE
x = odeint(model,x0,t)
# plot results
plt.plot(t,x)
plt.xlabel('time')
plt.ylabel('x(t)')
plt.show()
```

```python
#Experiment 5.3 Solve the second order differential equation

from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
def model(x,t):
     y = x[0]
     dy = x[1]  
     xdot = [[],[]]
     xdot[0] = dy
     xdot[1] = -2*dy-2*y+exp(-t)
     return xdot
time = np.linspace(0,2,100)
z1 = odeint(model,[0,0],time)
plt.plot(time,z1[:,0],'g:')
plt.plot(time,z1[:,1],'k-.')
plt.legend(['y (ODEINT)','dy/dt (ODEINT)'])
plt.xlabel('Time')
plt.show()
```

```python
#Experiment 5.2a: Solve for the current transient through an RC network (with RC = 3) t(for input =5)
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
RC=3
model = lambda x,t:((-x/RC)+(5/RC))
# initial condition
x0 = 1
# time points
t = np.linspace(0,20,10)
# solve ODE
x = odeint(model,x0,t)
# plot results
plt.plot(t,x)
plt.xlabel('time')
plt.ylabel('x(t)')
plt.show()


```

```python
#Experiment 5.2b:Solve for the current transient through an RC network (with RC = 3) t(for input signal (5e^−t)U(t))
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
RC=3
model = lambda x,t:((-x/RC)+((5/RC)*np.exp(-t)))
# initial condition
x0 = 1
# time points
t = np.linspace(0,2,10)
# solve ODE
x = odeint(model,x0,t)
# plot results
plt.plot(t,x)
plt.xlabel('time')
plt.ylabel('x(t)')
plt.show()
```
