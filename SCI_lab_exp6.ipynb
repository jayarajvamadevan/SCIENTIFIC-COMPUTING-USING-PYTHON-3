```python
#Experiment 6:Simple Data Visualization
#Experiment 6.1: Draw stem plots, line plots, box plots, bar plots and scatter plots with random data
# a) Stem plot:

from numpy import random
import matplotlib.pyplot as plt
x=random.randn(30)
plt.stem(x,use_line_collection = True)
plt.ylabel('some numbers')
plt.show()

```

```python
# b) Line plot:
from numpy import random
import matplotlib.pyplot as plt

x=random.randn(30)

plt.plot(x, color='red', marker='o')
plt.ylabel('some numbers')
plt.show()
```

```python
# c) Box plots:
from numpy import random
import matplotlib.pyplot as plt
x=random.randn(30)
plt.boxplot(x)
plt.ylabel('some numbers')
plt.show()
```

```python
# d) Bar plots:
from numpy import random
import matplotlib.pyplot as plt
x=random.randn(30)
y=range(0,30)
plt.bar(y,x,color ='purple',width = 0.8)
plt.ylabel('some numbers')
plt.show()
```

```python
# e) scatter plots
from numpy import random
import matplotlib.pyplot as plt
%matplotlib inline
x=random.randn(30)
y=range(0,30)
plt.scatter(y,x, color='b')
plt.ylabel('some numbers')
plt.show()
```

```python
#Experiment 6.2: Plot the histogram of a random data.
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(30)
y= np.random.normal(size=100)
plt.hist(y, density=True, bins=20)
plt.ylabel('Probability')
plt.xlabel('Data');
```

```python
#Experiment 6.3: Create legends in plots.

import numpy as np 
import matplotlib.pyplot as plt 
y1 = [1,2,3,4,5,6,7,8] 
y2 = [8,7,6,5,4,3,2,1]   
plt.plot(y1,'b') 
plt.plot(y2,'r') 
  
# Function add a legend   
plt.legend(["blue", "green"], loc ="upper center") 
plt.show() 
```

```python
#Experiment 6.4: Create legends in plots  another example.

import numpy as np 
import matplotlib.pyplot as plt 
x = np.arange(5) 
y1 = [1, 2, 3, 4, 5] 
y2 = [1, 4, 9, 16, 25] 
# Function to plot   
plt.plot(x, y1, '--b',label ='Numbers') 
plt.plot(x, y2,'r:',label ='Square of numbers') 
# Function add a legend   
plt.legend() 
 
plt.show() 
```

```python
#Experiment 6.4:Realize a vector t = [10, 10] with increment 0.01 as an array.
import numpy as np
x = np.linspace(-10,10,100)
plt.stem(x,use_line_collection = True)
plt.ylabel('some numbers')
plt.show()

```

```python
##Experiment 6.4: Implement and plot the functions f(t)=cost and f(t)=costcos5t+cos5t

import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0, 10, 0.01)
y1 = np.cos(t)
y2 = (np.cos(t)*np.cos(5*t))+np.cos(5*t)
plt.subplot(211)
plt.plot(t,y1,'r--')
plt.grid(True)
plt.subplot(212)
plt.plot(t,y2,'g:')
plt.grid(True)


```
