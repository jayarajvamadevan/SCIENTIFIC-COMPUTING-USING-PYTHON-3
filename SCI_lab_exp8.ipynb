```python
#8.1,2 Realize the Fourier series

import matplotlib.pyplot as plt
import numpy as np

N_max = 5
n_odds = np.arange(1,N_max,1)
T=10
xs = np.arange(0,100,0.01)
ys = []
for x in xs:
    sum_terms = []
    for n_odd in n_odds:
      sum_term = ((-1)**(n_odd+1))*(4/((2*n_odd-1)*np.pi))*np.cos((2*np.pi*(2*n_odd-1)*x)/T)    
      sum_terms.append(sum_term)
      y = sum(sum_terms)
    ys.append(y)

plt.plot(xs, ys)
plt.show()
```

```python
#8.4 Plot the first 3 or 4 terms on the same graphic window and understandhow the smooth sinusoids add up to a discontinous square function.
import matplotlib.pyplot as plt
import numpy as np

N =[3,4]
for N_max in N:
    n_odds = np.arange(1,N_max,1)
    T=20
    xs = np.arange(0,100,0.01)
    ys = []
    for x in xs:
        sum_terms = []
        for n_odd in n_odds:
            sum_term = ((-1)**(n_odd+1))*(4/((2*n_odd-1)*np.pi))*np.cos((2*np.pi*(2*n_odd-1)*x)/T)    
            sum_terms.append(sum_term)
            y = sum(sum_terms)
        ys.append(y)
    plt.plot(xs, ys,'r')
    plt.show()
```
